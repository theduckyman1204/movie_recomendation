```python


# Lấy danh sách tất cả thể loại (trừ "(no genres listed)")
all_genres = sorted(set(genre for sublist in df_movie["genres_list"] for genre in sublist if genre != "(no genres listed)"))

# Mã hóa one-hot và lưu dưới dạng array trong một cột duy nhất
df_movie["encoded"] = df_movie["genres_list"].apply(lambda x: np.array([1 if genre in x else 0 for genre in all_genres]))

# Hiển thị kết quả
df_movie
```

```python
# Kết hợp dữ liệu để lấy encoded tương ứng cho mỗi movieId
df_rating = df_rating.merge(df_movie[["movieId", "encoded"]], on="movieId", how="left")

df_rating
```

```python
df_rating['weight_encoding'] = df_rating.apply(lambda row : row['rating'] * row['encoded'], axis = 1)
df_rating
```

```python
df_user_profiles = df_rating.groupby("userId")["weight_encoding"].apply(lambda x: np.mean(np.stack(x), axis=0)).reset_index()
df_user_profiles
```
